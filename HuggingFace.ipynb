```python
from transformers import pipeline

# Load the Named Entity Recognition (NER) pipeline
ner = pipeline("ner", grouped_entities=True)  # Grouping entities together

# Example sentences
sentences = [
    "Elon Musk is the CEO of Tesla and SpaceX.",
    "The Eiffel Tower is located in Paris, France.",
    "Apple Inc. was founded by Steve Jobs, Steve Wozniak, and Ronald Wayne.",
    "The FIFA World Cup 2022 was held in Qatar.",
    "NASA launched the Artemis I mission to the Moon."
]

# Perform NER
for sentence in sentences:
    entities = ner(sentence)
    print(f"Sentence: {sentence}")
    for entity in entities:
        print(f"Entity: {entity['word']}, Type: {entity['entity_group']}, Confidence: {entity['score']:.2f}")
    print("\n")
```

```python
# Usage
import torch
from transformers import LlamaTokenizer, LlamaForCausalLM

tokenizer = LlamaTokenizer.from_pretrained('sarvamai/OpenHathi-7B-Hi-v0.1-Base')
model = LlamaForCausalLM.from_pretrained('sarvamai/OpenHathi-7B-Hi-v0.1-Base', torch_dtype=torch.bfloat16)

prompt = "मैं एक अच्छा हाथी हूँ"
inputs = tokenizer(prompt, return_tensors="pt")

# Generate
generate_ids = model.generate(inputs.input_ids, max_length=30)
tokenizer.batch_decode(generate_ids, skip_special_tokens=True, clean_up_tokenization_spaces=False)[0]
```
